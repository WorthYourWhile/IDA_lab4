```python
from transformers import pipeline
```

```python

question = "–Ø–∫–∏–π —Ñ–æ—Ä–º–∞—Ç –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –∑–∞–≤–¥–∞–Ω–Ω—è?"
context = """
–û–ø–∏—à—ñ—Ç—å, –±—É–¥—å –ª–∞—Å–∫–∞, —è–∫—É —Ç–µ–º–∞—Ç–∏–∫—É –º–∞–≥—ñ—Å—Ç–µ—Ä—Å—å–∫–æ—ó —Ä–æ–±–æ—Ç–∏ –≤–∏ –± —Ö–æ—Ç—ñ–ª–∏ –æ–±—Ä–∞—Ç–∏. –í –µ—Å–µ –¥–∞–π—Ç–µ –≤—ñ–¥–ø–æ–≤—ñ–¥—ñ –Ω–∞ –∑–∞–ø–∏—Ç–∞–Ω–Ω—è:
1. –ß–∏–º –æ–±–≥—Ä—É–Ω—Ç–æ–≤–∞–Ω–∏–π –≤–∏–±—ñ—Ä —Ç–µ–º–∏?
2. –ß–ò –º–∞—î—Ç–µ –≤–∏ –¥–æ—Ç–∏—á–Ω–∏–π –¥–æ—Å–≤—ñ–¥?
3. –Ø–∫—ñ –≤–∞—à—ñ –≤–∞—à—ñ –∫—Ä–∏—Ç–µ—Ä—ñ—ó –≤–∏–±–æ—Ä—É —Ç–µ–º–∏?
4. –Ø–∫ –≤–æ–Ω–∞ –ø–æ–≤'—è–∑–∞–Ω–∞ –∑ –Ω–∞–ø—Ä—è–º–∫–∞–º–∏ –Ω–∞—à–æ—ó –∫–∞—Ñ–µ–¥—Ä–∏?


–ö—Ä–∏—Ç–µ—Ä—ñ—ó –æ—Ü—ñ–Ω—é–≤–∞–Ω–Ω—è:
1. –ü–µ—Ä—Å–æ–Ω—ñ—Ñ—ñ–∫–æ–≤–∞–Ω—ñ –¥—É–º–∫–∏ (–Ω–∞–≤—ñ—Ç—å —è–∫—â–æ –¥–æ–ø–æ–º–∞–≥–∞—î –ø–∏—Å–∞—Ç–∏ –®–Ü, —è –º–∞—é –±–∞—á–∏—Ç–∏ —Ö—ñ–¥ –≤–∞—à–æ—ó, –∞ –Ω–µ –π–æ–≥–æ –¥—É–º–∫–∏) - 10 –±–∞–ª—ñ–≤
2. –ù–∞—è–≤–Ω—ñ—Å—Ç—å —á—ñ—Ç–∫–∏—Ö –∫—Ä–∏—Ç–µ—Ä—ñ—ó–≤ –≤–∏–±–æ—Ä—É - 5 –±–∞–ª—ñ–≤
3. –ù–∞—è–≤–Ω—ñ—Å—Ç—å "–∑–∞–¥—ñ–ª—É" (—î –≤–∂–µ —Ç–µ–º–∞, –∞–±–æ –Ω–∞—É–∫–æ–≤–∏–π –∫–µ—Ä—ñ–≤–Ω–∏–∫, –∞–±–æ –ø–µ–≤–Ω–∏–π –∑–∞–¥—ñ–ª –ø–æ —Ä–æ–±–æ—Ç—ñ) - 5 –±–∞–ª—ñ–≤
"""

answer = qa_model(question = question, context = context)
print("–í—ñ–¥–ø–æ–≤—ñ–¥—å:", answer['answer'])
```

```python
whos = unmasker("–õ—é–¥–∏–Ω–∞, —è–∫–∞ —Å—Ç–≤–æ—Ä–∏–ª–∞ –æ–¥–Ω–µ –∑ –Ω–∞–π–∫—Ä–∞—â–∏—Ö –∫–æ–º'—é–Ω—ñ—Ç—ñ —É –º–µ–∂–∞—Ö —É–∫—Ä–∞—ó–Ω—Å—å–∫–æ–≥–æ <mask> ü§°")
for who in whos:
    print(who['token_str'])
```
